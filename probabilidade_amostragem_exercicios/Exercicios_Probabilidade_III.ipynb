Os dados de uma pesquisa mostram algumas informações sobre o tempo de cirurgias para recostrução ACL em hospitais com alto volume de cirurgia. A partir dos dados foram calculados, o tempo médio de 129 minutos com um desvio padrão de 14 minutos.

a - Qual é a probabilidade de uma cirurgia ACL, em um hospital com alto volume de cirurgias, requerer um tempo maior do que dois desvios-padrão acima da média?

b - Qual é a probabilidade de uma cirurgia ACL, em um hospital com alto volume de cirurgias ser completada em menos de 100 minutos?

c - Em qual tempo a probabilidade de uma cirurgia ACL em um hospital com alto volume de cirurgias é igual a 0.95?

d - Se uma cirurgia requer 199 minutos, o que você conclui sobre o volume de tais cirurgias em um hospital? Explique.

```python
from scipy.stats import norm

# Média e desvio padrão dados
media = 129
desvio_padrao = 14

# Calculando o limite superior (média + 2 * desvio padrão)
limite_superior = media + 2 * desvio_padrao

# Calculando a probabilidade usando a função de distribuição cumulativa (1 - cdf)
probabilidade = 1 - norm.cdf(limite_superior, loc=media, scale=desvio_padrao)

print(f"A probabilidade de uma cirurgia ACL levar mais\ndo que dois desvios-padrão acima da média é de {probabilidade:.4f}")
```

```python
from scipy.stats import norm

# Definir os parâmetros da distribuição normal
tempo_medio = 129  # minutos
desvio_padrao = 14  # minutos

# Calcular a probabilidade usando a distribuição normal acumulada (CDF)
probabilidade = norm.cdf(100, tempo_medio, desvio_padrao)

# Exibir a probabilidade formatada como porcentagem
probabilidade_percentual = probabilidade * 100
print(f'A probabilidade de uma cirurgia ACL ser\ncompletada em menos de 100 minutos é de aproximadamente {probabilidade_percentual:.2f}%')
```

```python
from scipy.stats import norm

# Média e desvio padrão
media = 129
desvio_padrao = 14

# Calculando o Z-score para a probabilidade desejada
probabilidade = 0.95
z_score = norm.ppf(probabilidade)

# Calculando o tempo com base no Z-score
tempo = media + z_score * desvio_padrao

print(f"O tempo em que a probabilidade é\nigual a 0.95 é de aproximadamente {tempo:.2f} minutos.")
```

```python
# Definindo os valores
x = 199  # Tempo observado
media = 129  # Tempo médio
desvio_padrao = 14  # Desvio padrão

# Calculando o escore z
z = (x - media) / desvio_padrao

# Exibindo o escore z
print("Escore z:", z)
```

**QUESTAO 5**

Uma fábrica de carros sabe que os motores de sua fabricação têm duração normal
com média 150000 km e desvio-padrão de 5000 km. Qual a probabilidade de que um
carro, escolhido ao acaso, dos fabricados por essa firma, tenha um motor que dure:

(a) Menos de 170000 km?

(b) Entre 140000 km e 165000 km?

(c) Se a fábrica substitui o motor que apresenta duração inferior à garantia, qual deve ser esta garantia para que a porcentagem de motores substituídos seja inferior a 0,2%?

```python
from scipy.stats import norm

# Parâmetros da distribuição normal
media = 150000  # Média em km
desvio_padrao = 5000  # Desvio-padrão em km
```

```python
# (a) Probabilidade de um carro ter um motor que dure menos de 170000 km:
limite_superior_a = 170000
probabilidade_a = norm.cdf(limite_superior_a, loc=media, scale=desvio_padrao)
print(f"A probabilidade é de {probabilidade_a * 100:.2f}%")
```

```python
# (b) Probabilidade de um carro ter um motor que dure entre 140000 km e 165000 km:
limite_inferior_b = 140000
limite_superior_b = 165000
probabilidade_b = norm.cdf(limite_superior_b, loc=media, scale=desvio_padrao) - norm.cdf(limite_inferior_b, loc=media, scale=desvio_padrao)
print(f"A probabilidade é de {probabilidade_b * 100:.2f}%")
```

```python
# (c) Determinação da garantia necessária para que a porcentagem de motores substituídos seja inferior a 0,2%:

probabilidade_substituicao = 0.002  # 0,2% expresso como decimal

# Utilizando a inversa da função de distribuição acumulada (ppf) para determinar o valor correspondente à probabilidade
garantia = norm.ppf(probabilidade_substituicao, loc=media, scale=desvio_padrao)
print(f"A garantia deve ser de pelo menos {garantia:.2f} km")
```

**QUESTAO 8**

Uma urna tem 6 bolas brancas, 4 pretas e 5 azuis. Retiram-se 8 bolas com reposição. Qual a probabilidade de sair 4 bolas brancas, 2 pretas e 2 azuis?

```python
from math import factorial

# Número de bolas de cada cor
brancas = 6
pretas = 4
azuis = 5

# Número total de bolas na urna
total_bolas = brancas + pretas + azuis

# Número de retiradas desejadas
retiradas_desejadas = 8

# Número de bolas desejadas de cada cor nas retiradas
brancas_desejadas = 4
pretas_desejadas = 2
azuis_desejadas = 2

# Calculando as probabilidades de cada cor
prob_branca = brancas / total_bolas
prob_preta = pretas / total_bolas
prob_azul = azuis / total_bolas

# Calculando a probabilidade total do evento desejado
prob_evento_desejado = (prob_branca ** brancas_desejadas) * (prob_preta ** pretas_desejadas) * (prob_azul ** azuis_desejadas)

# Calculando o número total de maneiras de ocorrer o evento
maneiras_total = factorial(8)/(factorial(4)*factorial(2)*factorial(2))

# Calculando a probabilidade final
probabilidade_final = prob_evento_desejado * maneiras_total

print(f'A probabilidade de sair 4 bolas brancas, 2 pretas e 2 azuis é {probabilidade_final:.4f}')
```

```python
from scipy.stats import multinomial

# Número de bolas de cada cor
brancas = 6
pretas = 4
azuis = 5

# Número de retiradas desejadas
retiradas_desejadas = 8

# Número de bolas desejadas de cada cor nas retiradas
bolas_desejadas = [4, 2, 2]

# Probabilidades de cada cor
probabilidades = [brancas / (brancas + pretas + azuis),
                  pretas / (brancas + pretas + azuis),
                  azuis / (brancas + pretas + azuis)]

# Calculando a probabilidade usando multinomial.pmf
probabilidade = multinomial.pmf(bolas_desejadas, retiradas_desejadas, probabilidades)

print(f'A probabilidade de sair 4 bolas brancas, 2 pretas e 2 azuis é {probabilidade:.4f}')
```

**QUESTAO 9**

Suponha que a probabilidade de que você faça uma venda durante qualquer um dos telefonemas feitos é 0,19. Encontre a probabilidade de que você:

a) faça sua primeira venda durante a quinta ligação;

b) faça sua primeira venda durante a primeira, segunda ou terceira ligação;

c) não faça uma venda durante as três primeiras ligações.

```python
# Importando a função necessária da biblioteca geom
from scipy.stats import geom

# Definindo a probabilidade de sucesso (fazer uma venda)
p = 0.19
```

```python
# a) Probabilidade de fazer a primeira venda na quinta ligação
k_a = 5
prob_a = geom.pmf(k_a, p)
print(f"Probabilidade de fazer a primeira venda na quinta ligação: {prob_a:.4f}")
```

```python
# b) Probabilidade de fazer a primeira venda na primeira, segunda ou terceira ligação
k_b = [1, 2, 3]
prob_b = sum(geom.pmf(k, p) for k in k_b)
print(f"Probabilidade de fazer a primeira venda na primeira, segunda ou terceira ligação: {prob_b:.4f}")
```

```python
# b) Probabilidade de fazer a primeira venda na primeira, segunda ou terceira ligação
k_b = 3
prob_b = geom.cdf(k_b, p)
print(f"Probabilidade de fazer a primeira venda na primeira, segunda ou terceira ligação: {prob_b:.4f}")
```

```python
# c) Probabilidade de não fazer uma venda nas três primeiras ligações
prob_c = 1 - prob_b
print(f"Probabilidade de não fazer uma venda nas três primeiras ligações: {prob_c:.4f}")
```

**QUESTAO 10**

Um produtor de vidro descobre que 1 em cada 500 itens de vidro está torcido. Encontre a probabilidade de:

a) o primeiro item de vidro torcido ser o décimo item produzido;

b) o primeiro item de vidro torcido ser o segundo ou terceiro item produzido;

c) nenhum dos dez primeiros itens de vidro estar imperfeito.

```python
from scipy.stats import geom

# Definir a taxa de sucesso (item de vidro estar torcido a cada 500 itens)
p_success = 1 / 500
```

```python
# a) Probabilidade do primeiro item torcido ser o décimo item produzido
prob_a = geom.pmf(10, p_success)
print(f"Probabilidade de o primeiro item torcido ser o décimo item produzido: {prob_a:.6f}")
```

```python
# b) Probabilidade do primeiro item torcido ser o segundo ou terceiro item produzido
prob_b = geom.pmf(2, p_success) + geom.pmf(3, p_success)
print(f"Probabilidade de o primeiro item torcido ser o segundo ou terceiro item produzido: {prob_b:.6f}")
```

```python
# b) Probabilidade do primeiro item torcido ser o segundo ou terceiro item produzido
prob_b = geom.cdf(3, p_success) - geom.cdf(1, p_success)
print(f"Probabilidade de o primeiro item torcido ser o segundo ou terceiro item produzido: {prob_b:.6f}")
```

```python
# c) Probabilidade de nenhum dos dez primeiros itens de vidro estar imperfeito
prob_c = 1 - prob_a
print(f"Probabilidade de nenhum dos dez primeiros itens de vidro estar imperfeito: {prob_c:.6f}")
```
