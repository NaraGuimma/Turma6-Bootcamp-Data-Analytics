```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Gerando dados normalmente distribuídos
dados_normais = np.random.normal(loc=0, scale=1, size=1000)

print(len(dados_normais))

# Plotando o histograma dos dados
plt.hist(dados_normais, bins=30, density=True, alpha=0.6, color='blue')

# Plotando a função de densidade de probabilidade
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Distribuição Normal')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.show()
```

```python
from scipy.stats import poisson

# Gerando dados seguindo uma distribuição de Poisson
dados_poisson = poisson.rvs(mu=3, size=1000)

# Plotando o histograma dos dados
plt.hist(dados_poisson, bins=15, density=True, alpha=0.6, color='green')

# Plotando a função de massa de probabilidade
k = np.arange(0, 12)
p = poisson.pmf(k, mu=3)
plt.plot(k, p, 'k', linewidth=2)
plt.title('Distribuição de Poisson')
plt.xlabel('Valores')
plt.ylabel('Probabilidade')
plt.show()
```

```python
from scipy.stats import binom

# Gerando dados seguindo uma distribuição binomial
dados_binom = binom.rvs(n=10, p=0.5, size=1000)
print(dados_binom)

# Plotando o histograma dos dados
plt.hist(dados_binom, bins=10, density=True, alpha=0.6, color='orange')

# Plotando a função de massa de probabilidade
k = np.arange(0, 11)
p = binom.pmf(k, n=10, p=0.5)
plt.plot(k, p, 'k', linewidth=2)
plt.title('Distribuição Binomial')
plt.xlabel('Número de Sucessos')
plt.ylabel('Probabilidade')
plt.show()
```

```python
from scipy.stats import bernoulli

# Gerando dados seguindo uma distribuição de Bernoulli
dados_bernoulli = bernoulli.rvs(p=0.3, size=1000)

# Plotando o histograma dos dados
plt.hist(dados_bernoulli, bins=2, density=True, alpha=0.6, color='purple')

# Plotando a função de massa de probabilidade
k = [0, 1]
p = bernoulli.pmf(k, p=0.3)
plt.plot(k, p, 'k', linewidth=2, marker='o')
plt.title('Distribuição de Bernoulli')
plt.xlabel('Resultado')
plt.ylabel('Probabilidade')
plt.xticks(k, ['Fracasso', 'Sucesso'])
plt.show()
```

```python
from scipy.stats import multinomial

# Definindo os parâmetros da distribuição multinomial
n = 20  # número total de eventos
p = [0.2, 0.3, 0.5]  # probabilidades de ocorrência em cada categoria

# Gerando dados seguindo uma distribuição multinomial
dados_multinomial = multinomial.rvs(n=n, p=p, size=1000)

# Plotando o histograma dos dados para cada categoria
categorias = ['Categoria 1', 'Categoria 2', 'Categoria 3']
for i in range(len(p)):
    plt.hist(dados_multinomial[:, i], bins=10, density=True, alpha=0.6, label=categorias[i])

plt.legend()
plt.title('Distribuição Multinomial')
plt.xlabel('Número de Ocorrências')
plt.ylabel('Probabilidade')
plt.show()
```

```python
from scipy.stats import geom

# Gerando dados seguindo uma distribuição geométrica
dados_geometrica = geom.rvs(p=0.3, size=1000)

# Plotando o histograma dos dados
plt.hist(dados_geometrica, bins=20, density=True, alpha=0.6, color='pink')

# Plotando a função de massa de probabilidade
k = np.arange(1, 21)
p = geom.pmf(k, p=0.3)
plt.plot(k, p, 'k', linewidth=2, marker='o')
plt.title('Distribuição Geométrica')
plt.xlabel('Número de Tentativas até o Primeiro Sucesso')
plt.ylabel('Probabilidade')
plt.show()
```

**QUESTAO 1**

Os dados de uma pesquisa mostram algumas informações sobre o tempo de cirurgias para recostrução ACL em hospitais com alto volume de cirurgia. A partir dos dados foram calculados, o tempo médio de 129 minutos com um desvio padrão de 14 minutos.

a - Qual é a probabilidade de uma cirurgia ACL, em um hospital com alto volume de cirurgias, requerer um tempo maior do que dois desvios-padrão acima da média?

b - Qual é a probabilidade de uma cirurgia ACL, em um hospital com alto volume de cirurgias ser completada em menos de 100 minutos?

c - Em qual tempo a probabilidade de uma cirurgia ACL em um hospital com alto volume de cirurgias é igual a 0.95?

d - Se uma cirurgia requer 199 minutos, o que você conclui sobre o volume de tais cirurgias em um hospital? Explique.

```python
from scipy.stats import norm

# Média e desvio padrão dados
media = 129
desvio_padrao = 14

# Calculando o limite superior (média + 2 * desvio padrão)
limite_superior = media + 2 * desvio_padrao

# Calculando a probabilidade usando a função de distribuição cumulativa (1 - cdf)
probabilidade = 1 - norm.cdf(limite_superior, loc=media, scale=desvio_padrao)

print(f"A probabilidade de uma cirurgia ACL levar mais\ndo que dois desvios-padrão acima da média é de {probabilidade:.4f}")
```

```python
from scipy.stats import norm

# Definir os parâmetros da distribuição normal
tempo_medio = 129  # minutos
desvio_padrao = 14  # minutos

# Calcular a probabilidade usando a distribuição normal acumulada (CDF)
probabilidade = norm.cdf(100, tempo_medio, desvio_padrao)

# Exibir a probabilidade formatada como porcentagem
probabilidade_percentual = probabilidade * 100
print(f'A probabilidade de uma cirurgia ACL ser\ncompletada em menos de 100 minutos é de aproximadamente {probabilidade_percentual:.2f}%')
```

```python
from scipy.stats import norm

# Média e desvio padrão
media = 129
desvio_padrao = 14

# Calculando o Z-score para a probabilidade desejada
probabilidade = 0.95
z_score = norm.ppf(probabilidade)

# Calculando o tempo com base no Z-score
tempo = media + z_score * desvio_padrao

print(f"O tempo em que a probabilidade é\nigual a 0.95 é de aproximadamente {tempo:.2f} minutos.")
```

```python
# Definindo os valores
x = 199  # Tempo observado
media = 129  # Tempo médio
desvio_padrao = 14  # Desvio padrão

# Calculando o escore z
z = (x - media) / desvio_padrao

# Exibindo o escore z
print("Escore z:", z)
```

Relembrando:
O Z-score, também conhecido como escore z, é uma medida estatística que indica o quão longe um determinado valor está da média de um conjunto de dados em termos de desvios padrão. É uma forma de padronizar os dados e torná-los comparáveis, independentemente das unidades de medida originais.

**QUESTAO 2**

Suponha que foi constatado que a probabilidade de ter sangue tipo A é 0,4. Escolha 4 pessoas aleatoriamente e seja X o número com o tipo sanguíneo A.

a - Faça a distribuição com todas as probabilidades de x=0,1,2,3 ou 4

b - Faça um gráfico da distribuição do exercício 1

c - Calcule a probabilidade de exatamente duas pessoas terem sangue do tipo A.

d - Calcule a probabilidade de no máximo 3 pessoas terem sangue do tipo A.

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# Definindo os parâmetros do problema
p_tipo_A = 0.4  # Probabilidade de ter sangue tipo A
n_amostra = 4  # Número de pessoas escolhidas aleatoriamente

# a) Faça a distribuição com todas as probabilidades de x=0,1,2,3 ou 4
distribuicao = [binom.pmf(x, n_amostra, p_tipo_A) for x in range(n_amostra + 1)]
print("Distribuição de probabilidades:", distribuicao)
```

```python
# b) Faça um gráfico da distribuição
plt.bar(range(n_amostra + 1), distribuicao)
plt.xlabel('Número de pessoas com sangue tipo A')
plt.ylabel('Probabilidade')
plt.title('Distribuição de pessoas com sangue tipo A')
plt.show()
```

```python
# c) Calcule a probabilidade de exatamente duas pessoas terem sangue do tipo A
probabilidade_2_A = binom.pmf(2, n_amostra, p_tipo_A)
print("Probabilidade de exatamente duas pessoas terem sangue tipo A:", probabilidade_2_A)
```

```python
# d) Calcule a probabilidade de no máximo 3 pessoas terem sangue do tipo A
probabilidade_max_3_A = binom.cdf(3, n_amostra, p_tipo_A)
print("Probabilidade de no máximo 3 pessoas terem sangue tipo A:", probabilidade_max_3_A)
```

```python
# bonus, qual a probabilidade de se ter 4 pessoas com tipo sanguineo do tipo A

probabilidade_4 = 1 - probabilidade_max_3_A
print("Probabilidade de 4 pessoas terem sangue tipo A:", probabilidade_4)
```
